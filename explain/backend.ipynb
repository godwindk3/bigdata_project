### + app.py: Được xây dựng bằng framework fastapi.

### + Đây là phần "Application server" trong ảnh dưới.


### + Hàm read_stream(): Tạo ra một dòng data stream để lấy ra những thay đổi trong collection "live" ở database.


- Hàm watch() sẽ trả về đối tượng change_stream, đây là đối tượng chứa những thay đổi trong collection. <br>
<code>change_stream = message_collection.watch()</code>

- Dùng vòng for để duyệt tất cả những thay đổi trong đối tượng change_stream.
- Nếu như thay đổi được thực hiện bằng phép toán "insert" => Lấy ra thông tin của thay đổi đó và trả lại.

    

=> Những thay đổi này chính là những đoạn chat được gửi đi trên live video, hàm này sẽ track mỗi khi có tin nhắn mới được thêm vào collection "live" và trả lại những tin nhắn đó.
    

### + Các API:
- get("/stream"): Khi truy cập vào api này người dùng sẽ nhận được một dòng data stream đến từ những thay đổi trong collection "live" của database.

    

- get("video/{id}): Lấy ra thông tin của video (URL, tên video, ...)
