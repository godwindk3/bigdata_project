```python
from translate import Translator
```

```python
trans = Translator(from_lang="autodetect", to_lang="en")
```

```python
jap_sentence = "現実は拒絶され、シナプスは砕け散り、この世から追放される."
trans.translate(jap_sentence)
```

```python
italy_sentence = "vafanculo."
trans.translate(italy_sentence)
```

```python
pinoy_sentence = "Gusto ko ng adobo, paborito ko iyon."
trans.translate(pinoy_sentence)
```

```python
china_sentence = "你是中国人吗"
trans.translate(china_sentence)
```
